```python
import pandas as pd
df = pd.read_csv("browser_rankings_data.csv",skiprows=1)
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df['Short Description']=df['Short Description'].astype('category').cat.codes
df['Long Description']=df['Long Description'].astype('category').cat.codes
df['Keyword']=df['Keyword'].astype('category').cat.codes
df['App ID']=df['App ID'].astype('category').cat.codes
corr = df.corr()
```

```python
sns.pairplot(corr)
```

```python
sns.heatmap(corr,annot=True)
```

### Is there any co-relation between short description, long description and ranking? Does the placement of keyword (for example - using a keyword in the first 10 words - have any co-relation with the ranking)?


Answer: From the above heatmap and pairplots we can clearly see:
                    
                    1. The shot and long description follows 50% co-relation with each other, and the negative sign can be justified by watcing the pairplot which follows a negative log function curve.
                    
                    2. The short description and rank share a negative co-relation of just 22% can be justified by saying they do not co-relate significantly.
                    
                    3. Long description and rank similarly shows a positive co-relation of 34%, can be said that they are somewhat related to eachother.
                    
                    4. Keyword and rank show nearly neglisible co-relation. This it wont affect anyway by the placement of keyword.

### Does APP ID (Also known as package name) play any role in ranking?

Answer: App ID is highly co-related with rank by 53%. From pairplot, it can be observed, they follow a linear relation. So, AppID can be a major factor in predicting rank.
